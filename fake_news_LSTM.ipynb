```python
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
url = 'https://raw.githubusercontent.com/Swapnita04/Fake-News-Detection/master/train.csv'
df= pd.read_csv(url)
```

```python
df.head()
```

```python
# Drop NAN values
df = df.dropna()
```

```python
# Get the Independant features
x = df.drop('label',axis=1)
```

```python
# Get the Dependant feature
y = df['label']
```

```python
x.shape
```

```python
y.shape
```

```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
## Vocabulary Size
voc_size = 5000
```

```python
## OneHot Representation
```

```python
messages = x.copy()
```

```python
messages.reset_index(inplace = True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
# DATA PREPROCESSING
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0,len(messages)):
    print(i)
    review = re.sub('[^a-zA-Z]',' ',messages['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if word not in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
onehot_rep = [one_hot(words,voc_size)for words in corpus] # giving index based on voc_size
onehot_rep
```

```python
## Embedding Representation
```

```python
sent_length = 20
embedded_docs = pad_sequences(onehot_rep, padding ='pre', maxlen = sent_length)
print(embedded_docs)
```

```python
len(embedded_docs)
```

```python
## Creating Model
```

```python
#from nltk import metrics
embedding_vector_features = 40
model = Sequential()
model.add(Embedding(voc_size, embedding_vector_features, input_length = sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
print(model.summary)
```

```python
len(embedded_docs),y.shape
```

```python
import numpy as np
x_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
x_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_final, y_final, test_size =0.33, random_state =42)
```

```python
# Model training
```

```python
model.fit(x_train,y_train,validation_data = (x_test,y_test),epochs =10, batch_size =64)
```
